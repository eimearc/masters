```python
CSV_DIR='csv'

df_vulkan = pd.read_csv(path.join(CSV_DIR,'time_vulkan_final.csv'))
df_vulkan_new = pd.read_csv(path.join(CSV_DIR,'time_vulkan.csv'))
df_gl = pd.read_csv(path.join(CSV_DIR,'time_gl_final.csv'))

for df in [df_vulkan, df_gl, df_vulkan_new]:
    z=np.abs(stats.zscore(df.frame))
    df.drop(df.index[np.where(z>3)[0]], inplace=True) # Make sure to have inplace=True

def plots(name,df):
    global df_vulkan
    global df_gl
    max_frame=max([df_vulkan.frame.max(), df_vulkan_new.frame.max(),df_gl.frame.max()])

    x_label='time (ms)'
    y_label='count'
    
    fig,ax=plt.subplots(1,4,figsize=(20,5))
    fig.suptitle('{} Frame Render Time (ms) - {} Vertices'.format(name,df.vertices.iloc[0]))
    for i in range(1,5):
        df[df.threads==i].hist('frame',bins=100,ax=ax[i-1],alpha=0.5,range=[0,max_frame])
        ax[i-1].set_xlabel(x_label)
        ax[i-1].set_ylabel(y_label)
        ax[i-1].set_title('{} threads'.format(i))
    plt.show()

#     fig,ax=plt.subplots(1,4,figsize=(20,5))
#     fig.suptitle('{} VBO Update Time (ms) - {} Vertices'.format(name,df.vertices.iloc[0]))
#     for i in range(1,5):
#         df[df.threads==i].hist('updateVBO',bins=100,ax=ax[i-1],alpha=0.5,range=[0,df.updateVBO.max()])
#         ax[i-1].set_xlabel(x_label)
#         ax[i-1].set_ylabel(y_label)
#         ax[i-1].set_title('{} threads'.format(i))
#     plt.show()

m = [df_vulkan.frame.max(), df_vulkan.updateVBO.max(),df_gl.frame.max(), df_gl.updateVBO.max()]

maximum=max(m)

plots('Vulkan',df_vulkan)
plots('Vulkan New',df_vulkan_new)
plots('OpenGL',df_gl)
```

```python
CSV_DIR='csv'
FILE='run01.csv'

print(path.join(CSV_DIR,'vulkan',FILE))
df_vulkan = pd.read_csv(path.join(CSV_DIR,'vulkan',FILE))
df_vulkan_startup = pd.read_csv(path.join(CSV_DIR,'vulkan','run03.csv'))
z=np.abs(stats.zscore(df_vulkan_startup.startup))
df_vulkan_startup.drop(df_vulkan_startup.index[np.where(z>3)[0]], inplace=True) # Make sure to have inplace=True
df_gl = pd.read_csv(path.join(CSV_DIR,'gl',FILE))

for df in [df_vulkan, df_gl]:
    z=np.abs(stats.zscore(df.frame))
    df.drop(df.index[np.where(z>3)[0]], inplace=True) # Make sure to have inplace=True

def gl_vulkan_compare():
    global df_vulkan
    global df_gl
    max_frame=max([df_vulkan.frame.max(),df_gl.frame.max()])
    
    fig,ax=plt.subplots()
    ax.set_title('Frame Render Time (ms) - {} Vertices'.format(df_vulkan.vertices.iloc[0]))
    ax.hist(df_gl.frame,bins=100,alpha=0.75,range=[0,max_frame],label='OpenGL')    
    ax.hist(df_vulkan.frame,bins=100,alpha=0.75,range=[0,max_frame],label='Vulkan')
    ax.legend(loc='upper right')
    plt.xlabel('time (ms)')
    plt.ylabel('count')
    
    plt.show()
    
def threads(df, n, subtitle, minimum=0, maximum=100):
    x_label='time (ms)'
    y_label='count'
    
    fig,ax=plt.subplots(1,4,figsize=(20,5))
    fig.suptitle('Vulkan {} Time (ms) - {} Vertices'.format(subtitle, df.vertices.iloc[0]))
    for i in range(1,5):
        df[df.threads==i].hist(n,bins=100,ax=ax[i-1],range=[minimum,maximum])
        ax[i-1].set_xlabel(x_label)
        ax[i-1].set_ylabel(y_label)
        ax[i-1].set_title('{} threads'.format(i))
    plt.show()

#gl_vulkan_compare()
threads(df_vulkan, 'frame', 'Frame Render', 0, df_vulkan.frame.max())
threads(df_vulkan_startup, 'startup', 'Startup', df_vulkan_startup.startup.min(), df_vulkan_startup.startup.max())
print(df_vulkan_startup.describe())
```
